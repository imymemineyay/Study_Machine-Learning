데이터프레임에서 데이터를 행 단위로 추출하려면 loc, iloc 속성을 사용해야함

### 데이터프레임 만들기

: 딕셔너리를 DataFrame클래스에 전달해야 함 <br>
⚠︎ 딕셔너리는 키와 값으로 이루어진 한쌍의 데이터들의 순서를 보장하지 않음 <br>
→ 순서가 보장된 딕셔너리를 전달하려면 OrderedDict 사용해야 함

: 인덱스를 따로 지정하지 않으면 인덱스를 0부터 자동으로 생성

: 인덱스를 따로 지정하려면 index 인자에 리스트 전달 

: columns인자를 사용하면 데이터프레임의 **열 순서 지정 가능**


```python
scientists = pd.DataFrame({
    'Name': ['Rosaline Franklin','William Gosset'],
    'Occupation' : ['Chemist','Statistician'],
    'Born': ['1920-07-25','1876-06-13'],
    'Died':['1958-04-16','1937-10-16'],
    'Age': [37, 61]
})

print(scientists)

print('-'*70)


scientists = pd.DataFrame(
    data = {'Name': ['Rosaline Franklin','William Gosset'],
    'Occupation' : ['Chemist','Statistician'],
    'Born': ['1920-07-25','1876-06-13'],
    'Died':['1958-04-16','1937-10-16'],
    'Age': [37, 61]},
    index = ['Rosaline Franklin', 'William Gosset'],
    columns = ['Name', 'Occupation', 'Born','Died','Age'])

print(scientists)
```

```python
from collections import OrderedDict # 순서가 보장된 딕셔너리 전달

scientists = pd.DataFrame(OrderedDict([
    ('Name', ['Rosaline Franklin','William Gosset']),
    ('Occupation', ['Chemist','Statistician']),
    ('Born', ['1920-07-25','1876-06-13']),
    ('Died', ['1958-04-16','1937-10-16']),
    ('Age', [37, 61]),    
])
)

print(scientists)
```

### 시리즈와 브로드캐스팅

브로드캐스팅 : 시리즈나 데이터프레임에 있는 모든 데이터에 대해 한 번에 연산하는 것

#### 박스 그래프

: 이산형 변수와 연속형 변수를 함께 사용하는 그래프

이산형 변수 : 명확하게 구분되는 값  ex) 남, 여

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
              tips[tips['sex'] == 'Male']['tip']],
              labels = ['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

### 다변량 그래프


: 3개 이상의 변수를 사용한 그래프

Why 다변량 그래프?

→ 3 개 이상의 변수를 사용하는 다변량 그래프는 적재적소에 맞는 그래프 요소를 추가하여 그래프를 표현해야 함

#### 산점도 그래프

앞서 실습한 산점도 그래프에 성별을 추가할 경우 색상을 통해 구분해야함 

그런데 문자열은 산점도 그래프의 색상을 지정하는 값으로 사용할 수 없기 때문에 0,1 과 같은 정수로 지정해줘야 함

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex) # recode_sex함수를 브로드캐스팅하기 위해 apply 메서드 사용
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size'] * 10, # 점의 사이즈
    c = tips['sex_color'], # 점의 컬러
    alpha = 0.3
)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## 4-5 seaborn 라이브러리로 그래프 스타일 설정하기

set_style 메서드를 사용하여 그래프의 스타일을 변경 가능

```python
fig,ax = plt.subplots()
ax = sns.violinplot(x='time',y='total_bill', hue='sex', data=tips, split=True)
```

```python
sns.set_style('whitegrid')
fig,ax = plt.subplots()
ax = sns.violinplot(x='time',y='total_bill',hue='sex',data=tips, split=True)
```

```python
fig = plt.figure()

seaborn_styles = ['darkgrid','whitegrid','dark','white','ticks']

for idx, style in enumerate(seaborn_styles) :
    plot_position = idx +1 
    with sns.axes_style(style):
        ax = fig.add_subplot(2,3,plot_position)
        violin = sns.violinplot(x='time',y='total_bill',data=tips,ax=ax)
        violin.set_title(style)

fig.tight_layout()
```
